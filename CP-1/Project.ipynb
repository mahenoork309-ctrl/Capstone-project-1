```python
# Fill missing numerical values with median
df['Resolution_Time_Hours'].fillna(df['Resolution_Time_Hours'].median(), inplace=True)
df['Satisfaction_Rating'].fillna(df['Satisfaction_Rating'].median(), inplace=True)

# Fill missing categorical values with mode
df['Assigned_Department'].fillna(df['Assigned_Department'].mode()[0], inplace=True)
df['Severity_Level'].fillna(df['Severity_Level'].mode()[0], inplace=True)
```

```python
# Correct Issue_Type
df['Issue_Type'] = df['Issue_Type'].replace({
    "Garbge": "Garbage",
    "Watr": "Water",
    "WiFiIssu": "WiFi Issue",
    "Pothl": "Pothole",
    "Trafic": "Traffic",
    "Stret Light": "Street Light",
    "Pollutn": "Pollution"})

# Correct Severity_Level
df['Severity_Level'] = df['Severity_Level'].replace({
    "Lo":"Low",
    "Med":"Medium",
    "Hi":"High",
    "Critcal":"Critical"})

# Correct Assigned_Department
df['Assigned_Department'] = df['Assigned_Department'].replace({
    "Watr Dept": "Water Dept",
    "Elect Dept": "Electricity Dept",
    "Road Dept": "Roads Dept",
    "SmartWifiDept": "Smart Wifi Dept",
    "Clean Dept": "Cleanliness Dept"})
```

```python
df.drop_duplicates(inplace=True)
```
